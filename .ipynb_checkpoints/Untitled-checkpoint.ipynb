```python
import cv2
import matplotlib.pyplot as plt
import pylab
from moviepy.editor import VideoFileClip
import numpy as np
from math import sqrt
from skimage import data
from skimage.feature import blob_dog, blob_log, blob_doh
from skimage.color import rgb2gray


```

```python
clip = VideoFileClip('978808.mp4')
d=0
kernel_size = 3;
scale = 1;
delta = 0;
fourcc = cv2.VideoWriter_fourcc('M','J','P','G')      #OpenCv writing file
out = cv2.VideoWriter('Detected_Lane3.avi',fourcc, 15.0, (596, 336)) # intialise output file


for frame in clip.iter_frames():
    if d==10:
        print(d)
        
        img= frame[:,:,2]
        plt.imshow(img,cmap='hot')
        plt.show()
        gray = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
        plt.imshow(gray,cmap='gray')
        plt.show()
        kernel = np.ones((9,9),np.float32)
        dst = cv2.blur(img,(7,7))
        plt.imshow(dst,cmap='gray')
        plt.show()
        dst=cv2.inRange(dst,0,130)
        plt.imshow(dst,cmap='gray')
        plt.show()
        
#         l=cv2.Laplacian( dst, ddepth=cv2.CV_16S ,scale=.1);
#         print(l)
#         plt.imshow(l,cmap='gray')
#         plt.show()
        
#         abs_dst=np.zeros_like(l)
#         abs_dst=cv2.convertScaleAbs( l, abs_dst );
#         plt.imshow(abs_dst,cmap='gray')
#         plt.show()
        blobs_log = blob_log(dst, max_sigma=50,num_sigma=20, threshold=.1)
       
        blobs_log[:, 2] = blobs_log[:, 2] * sqrt(2)
#         print(blobs_log[0])
        
#         blobs_list = [blobs_log, blobs_dog, blobs_doh]
        color =  'red'
        title = 'Laplacian of Gaussian'
#         sequence = zip(blobs_log, color, titles)
        
        i=frame
        
        fig, axes = plt.subplots(1)
        ax = axes
        ax.set_title(title)
        ax.imshow(frame,interpolation='nearest')
        print(len(blobs_log))
        for y,x,r in blobs_log:
            cv2.circle(i,(int(x),int(y)), int(r), (0,0,255), -1)

        
            c = plt.Circle((x, y), r, color=color, linewidth=1, fill=False)
            
            ax.add_patch(c)
        ax.set_axis_off()
        plt.tight_layout()
        plt.show()       
        plt.imshow(i)
        plt.show()
        out.write(i)


        
    
    d+=1
    
259
```
