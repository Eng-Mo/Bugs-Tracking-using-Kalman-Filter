```python
def blob_detection(image):
    img= image[:,:,2]
#     plt.imshow(img,cmap='hot')
#     plt.show()
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
#     plt.imshow(gray,cmap='gray')
#     plt.show()
    kernel = np.ones((9,9),np.float32)
    dst = cv2.blur(img,(7,7))
#     plt.imshow(dst,cmap='gray')
#     plt.show()
    dst=cv2.inRange(dst,0,130)
#     plt.imshow(dst,cmap='gray')
#     plt.show()

#         l=cv2.Laplacian( dst, ddepth=cv2.CV_16S ,scale=.1);
#         print(l)
#         plt.imshow(l,cmap='gray')
#         plt.show()

#         abs_dst=np.zeros_like(l)
#         abs_dst=cv2.convertScaleAbs( l, abs_dst );
#         plt.imshow(abs_dst,cmap='gray')
#         plt.show()
    blobs_log = blob_log(dst, max_sigma=30,num_sigma=20, threshold=.1)
#     print(len(blobs_log))

    blobs_log[:, 2] = blobs_log[:, 2] * sqrt(2)
#     print(blobs_log[0])
#     print(blobs_log.shape)
    measurments=blobs_log[:,:2]
#     print(measurments[0])

#         blobs_list = [blobs_log, blobs_dog, blobs_doh]
    color =  'red'
    title = 'Laplacian of Gaussian'
#         sequence = zip(blobs_log, color, titles)

    i=image

    fig, axes = plt.subplots(1)
    ax = axes
    ax.set_title(title)
    ax.imshow(image,interpolation='nearest')
    
    for y,x,r in blobs_log:
        cv2.circle(i,(int(x),int(y)), int(r), (0,0,255), -1)


        c = plt.Circle((x, y), r, color=color, linewidth=1, fill=False)

#         ax.add_patch(c)
#         ax.set_axis_off()
#         plt.tight_layout()
#         plt.show()       
#     plt.imshow(i)
#     plt.show()
#         out.write(i)
    return measurments
```

```python
def rondamColor():
    color_id= np.zeros([2000,3], dtype=np.int32)
    for color in range (0,len(color_id)):
        color_id[color,:]= (randint(0,255),randint(0,255),randint(0,255) )
        
        
    print('color ids= ',color_id)
    return color_id
    
```

```python
np.set_printoptions(suppress=True)
clip = VideoFileClip('978808.mp4')
d=0
kernel_size = 3;
scale = 1;
delta = 0;
fourcc = cv2.VideoWriter_fourcc('M','J','P','G')      #OpenCv writing file
out = cv2.VideoWriter('Detected_Lane3.avi',fourcc, 15.0, (596, 336)) # intialise output file
first_frame=clip.get_frame(0)
plt.title('frame')
plt.imshow(first_frame)
plt.show()
colors=rondamColor()
#################################################
dt=1 #Kalman step process
s_frame=6 #starting frame
u=0 # magnitude accelration start
acc_nois= 1 # accelration noise
measurement_noise=.1
ez=np.matrix([[measurement_noise, 0],[0, measurement_noise]])
ex= np.matrix([[dt**4/4, 0, dt**3/2, 0],
              [0, dt**4/4, 0, dt**3/2],
              [dt**3/2, 0, dt**2, 0],
              [0, dt**3/2, 0, dt**2]])*measurement_noise**2
p=ex

a=np.matrix([[1, 0, dt, 0],
           [0, 1, 0, dt],
           [0, 0, 1, 0],
           [0, 0, 0, 1]])  #starte update matrix
b= np.matrix([[(dt**2)/2],
            [(dt**2)/2],
            [dt],
            [dt]])

c=np.matrix([[1, 0, 0, 0],
           [0, 1, 0, 0]])

    
#################################################

##Kalman Pramaters

#Q_loc_meas = []
# %% initize estimation variables for two dimensions
# Q= [X{S_frame} Y{S_frame} zeros(length(X{S_frame}),1) zeros(length(X{S_frame}),1)]'
#  Q_estimate = [nan(4,2000)];
# Q_estimate(:,1:size(Q,2)) = Q;  %estimate of initial location estimation of where the flies are(what we are updating)
# Q_loc_estimateY = nan(2000); %  position estimate
# Q_loc_estimateX= nan(2000); %  position estimate
# P_estimate = P;  %covariance estimator
# strk_trks = zeros(1,2000);  %counter of how many strikes a track has gotten
# nD = size(X{S_frame},1); %initize number of detections
# nF =  find(isnan(Q_estimate(1,:))==1,1)-1 ; %initize number of track estimates
s_frame=6
# q=[]
q_estimate=np.zeros([2000,4])

q=blob_detection(first_frame)

# print(q.shape)
# print(len(q_estimate))
for i in range (0,len(q)):

    q_estimate[i,:2]=q[i,:2]
#             q_estimate[i,1]=q[i,1]

# print(q[0,:])

            

# print(q_estimate[0,:])
frame_track=np.zeros_like(first_frame)

nt=np.count_nonzero(q_estimate[:,0])
print('number tracking=',nt)
            

for frame in clip.iter_frames():
    print(d)
    q=[]
#     plt.imshow(frame)
#     plt.show()
    nt=np.count_nonzero(q_estimate[:,0])
    meas=blob_detection(frame)
    for x,y in meas:
        q.append([x,y])
    q=np.asarray(q)
#     plt.imshow(frame)
#     plt.show()
    
    if ((d>s_frame) |(d<200)):
        
        
#         print('measr shape=',q.shape)
#         print(len(q_estimate))
        nd=len(q)
        print('number detection=%d',nd)
        for tracke in range (0,nt):
#             print(b.shape)
#             print(q_estimate[tracke,:])
#             kalman=np.dot(a,q_estimate[tracke,:] )
#             print('kalamn=',kalman)
#             ykalman=b.getT() * u
#             print('ykalamn=',ykalman)
#             xkalman=kalman+ykalman
#             print('xkalamn=',xkalman)
            
            
            q_estimate[tracke,:] = np.dot(a, q_estimate[tracke,:] ) + b.T * u;
#             print('kalamn final= ', q_estimate[tracke,:])
            
            
        p =a*p*a.T + ex;
#             print('p==',p)
        k = p * c.T *inv(c*p*c.T+ez) 
#         print('kalman gain=', k)
            
       
            
            
 
            
        X=cdist(q_estimate[0:nt,0:2],q[:,0:2],'euclidean')
        
#         print('cost matrix= ', X)
        estimate_ind, meas_ind = linear_sum_assignment(X)
#         print('estimate undex= ',estimate_ind)
#         print('measr index=', meas_ind)
#         print('estimated =', q_estimate[meas_ind[:]])
#         print('detections =', q[meas_ind[:]])
       
        
        for i in range (0,len(estimate_ind)):
#             print('measurment= ', q[meas_ind[i],:])
#             print('estimated= ',q_estimate[estimate_ind[i],:])
            y= np.matrix(q[meas_ind[i],:]).T -  c*np.matrix(q_estimate[estimate_ind[i],:]).T
            
            
           

            q_estimate[estimate_ind[i],:]= np.matrix(np.matrix(q_estimate[estimate_ind[i],:]).T+k*y).T
#             print('new estimates= ',q_estimate[estimate_ind[i],:])
            q_estimate[estimate_ind[i],0:2]=np.ceil(q_estimate[estimate_ind[i],0:2])
#             print('new estimates= ',q_estimate[estimate_ind[i],:])
            
            
#         P =  (eye(4)-K*C)*P; 
        p=(np.eye(4)-k*c)*p
#         print('co variance= ',p)
        
        for i, point in enumerate(q_estimate[0:nt,0:2]):
            print('x and y =' ,point)
            print('color=',colors[i] )
            color= np.asarray((colors[i,0],colors[i,1],colors[i,2]),dtype=np.float)
            
            cv2.circle(frame_track,(int(point[1]),int(point[0])), 2, color , -1)
        for y,x in q[0:nd,0:2]:
            cv2.circle(frame,(int(x),int(y)), 2, (255,255,255), -1)
            
        
        frame=cv2.addWeighted(frame,.5,frame_track,.5,0)
#         plt.imshow(frame)
#         plt.show()
            
        out.write(frame)
        

           
    



        
    
    d+=1
    if d ==100:
        out.release()
        break
    
    
```
